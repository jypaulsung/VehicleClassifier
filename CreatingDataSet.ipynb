```python
from imutils import paths
import argparse
import requests
import cv2
import os
import urllib.request

url_list = "C:/Users/jungy/OneDrive/KHU/Software Convergence 101/sedanURLs.txt"
output_folder = "C:/Users/jungy/OneDrive/KHU/Software Convergence 101/images/sedan"

urls = open(url_list).read().strip().split("\n")
total = 0
for url in urls:
    try:
        r = requests.get(url, timeout=30)
        p = output_folder + "/" + str(total) + ".jpg"
        f = open(p, "wb")
        f.write(r.content)
        f.close()
        print("[INFO] downloaded: {}".format(p))
        total += 1
    except:
        print("[INFO] error downloading {}...skipping".format(p))


for imagePath in paths.list_images(output_folder):
    delete = False
    try:
        image = cv2.imread(imagePath)
        if image is None:
            delete = True
    except:
        print("Except")
        delete = True
    if delete:
        print("[INFO] deleting {}".format(imagePath))
        os.remove(imagePath)




    
```
